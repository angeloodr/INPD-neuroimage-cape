```python
#OPTIONS TO CREATE ROIS - TWO
# Create a Spherical ROI
import nibabel as nb
from nilearn.maskers import NiftiMasker
from templateflow.api import get
atlas_file = get('MNI152NLin2009cAsym',
resolution=2,
atlas='HOSPA',
desc='th25'
)
atlas = nb.load(atlas_file)

# Step 1: Create a zero volume with the same shape as the atlas
zero_volume = np.zeros(atlas.shape, dtype=np.float32)

# Step 2: Use the affine matrix from the atlas
affine = atlas.affine

# Step 3: Define the center and radius of the sphere
center = (32, 32, 32)  # Example coordinates; change as needed
radius = 10  # Example radius; change as needed

# Step 4: Create a meshgrid for the volume
x = np.arange(0, zero_volume.shape[0])
y = np.arange(0, zero_volume.shape[1])
z = np.arange(0, zero_volume.shape[2])
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

# Step 5: Calculate the distance from the center for each voxel
distance = np.sqrt((X - center[0])**2 + (Y - center[1])**2 + (Z - center[2])**2)

# Step 6: Create the spherical ROI by setting voxels within the radius to 1
zero_volume[distance <= radius] = 1

# Step 7: Create a NIfTI image using the modified zero_volume
roi_image = nib.Nifti1Image(zero_volume, affine)

# Step 8: Save the ROI image to a file (optional)
nib.save(roi_image, 'spherical_roi.nii')

```
